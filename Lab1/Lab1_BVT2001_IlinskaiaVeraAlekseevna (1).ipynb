### Задание №3

```python
from random import randint

#создание рандомной матрицы для задания
m = int (input("Введите m = ") or 50) 
n = int (input("Введите n = ") or 50)
min_limit = int (input("Введите минимальное значение генерируемого числа = ") 
                 or -250)
max_limit = int (input("Введите максимальное значение генерируемого числа = ") 
                 or 1000)

arr = []
for i in range (0,m): 
    arr.append([]) 
    for j in range (0,n): 
        arr[i].append(randint(min_limit, max_limit))

print()

print("Исходная матрица: \n")
for s in arr:
    print(s)

    
print() #пробел между матрицами


#функция сортировки выбором
#сравнение и обмен фиксированного a элемента с элементами от a+1 до длинны строки
def Selection_Sort (arr1):
    for i in range (0, len(arr1)): # len(arr) == m
         for a in range (0, len(arr1[i])-1): # len(arr[i]) == n
            for b in range(a+1, len(arr1[i])): 
                if (arr1[i][a] > arr1[i][b]):
                    arr1[i][a], arr1[i][b] = arr1[i][b], arr1[i][a]
    return arr1

print("Матрица, сортированная с помощью метода сортировки выбором: \n")
arr_selection = Selection_Sort(arr)
for s in arr_selection:
    print(s)

    
print() #пробел между матрицами


#функция сортировки вставкой
#сравнение и обмен фиксированного а элемента с элементами a-1 (справа налево)
def Insertion_Sort (arr2):
    for i in range(0, len(arr2)):
        for a in range(0, len(arr2[i])):
            fics = arr2[i][a]
            pos = a
            while (pos>0 and arr2[i][pos-1] > fics):
                arr2[i][pos] = arr2[i][pos-1]
                pos-=1
                arr2[i][pos] = fics
    return arr2

print("Матрица, сортированная с помощью метода сортировки вставкой: \n")
arr_insertion = Insertion_Sort(arr)
for s in arr_insertion:
    print(s)
    
    
print() #пробел между матрицами


#функция сортировки обменом (пузырьком)
#сравнение пар соседних элементов и обмен
def Exchange_Sort (arr3):
    for i in range(0, len(arr3)):
        for run in range(0, len(arr3[i])-1): #для многоразовой сортировки одной строки по парам
            for a in range (len(arr3[i])-1): #len-1, так как у последнего элемента нету пары a+1
                if (arr3[i][a] > arr3[i][a+1]): #пузырьком является последнее число обмена
                    arr3[i][a], arr3[i][a+1] = arr3[i][a+1], arr3[i][a]
    return arr3

print("Матрица, сортированная с помощью метода сортировки обменом: \n")
arr_exchange = Exchange_Sort(arr)
for s in arr_exchange:
    print(s)

    
print() #пробел между матрицами


#функция сортировки Шелла (улучшенная сортировка вставкой)
#сортировка подмассивов и сравнение элементов через определённый диапазон
def Shell_Sort (arr4):
    for i in range (0, len(arr4)):
        m = len(arr4[i])
        interval = m//2 #с помощью интвервала разделяем массив на подмассивы
                        #их элементы мы будем между собой сравнивать
        while (interval > 0):
            for a in range(interval, m):
                fics = arr4[i][a] #берём элемент второго подмассива
                pos = a #сохраняем позицию элемента второго подмассива
                while (pos >= interval and arr4[i][pos-interval] > fics):
                    #arr[i][a-interval] > fics - сравнение элемента первого массива со вторым 
                    arr4[i][pos] = arr4[i][pos-interval]
                    pos -= interval
                    arr4[i][pos] = fics
            interval //= 2 #например, если 8, то разбиваем на 4, затем на 2 и 1, и сравниваем элементы
    return arr4

print("Матрица, сортированная с помощью метода сортировки Шелла: \n")
arr_shell = Shell_Sort(arr)
for s in arr_shell:
    print(s)


print() #пробел между матрицами

   
#функция быстрой сортировки 
#сортировка, основанная на рекурсии и выборкой опорного
#элемента на каждом рекурсивном этапе
final_fastsort_list5 = [];
def Fast_Sort (str5):
    less = []
    equal = [] 
    more = []
    if (len(str5) > 1):
        opora = str5[0]
        for x in str5:
            if (x < opora):
                less.append(x)
            elif (x == opora):
                equal.append(x)
            elif (x > opora):
                more.append(x)
        return Fast_Sort(less) + equal + Fast_Sort(more)
    else:
        return str5
    
print("Матрица, сортированная с помощью метода быстрой сортировки: \n")
for i in range(len(arr)):
    final_fastsort_list5.append(Fast_Sort(arr[i]))
for j in range (len(final_fastsort_list5)):
    print(final_fastsort_list5[j])
```
