```python
t = input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö: ")
t = int(t)

for i in range(t):
  n = input("\n–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω—É –ø—É—Ç–∏ —Ä–æ–±–æ—Ç–∞: ")
  n = int(n)
  s = input("–í–≤–µ–¥–∏—Ç–µ –ø—É—Ç—å —Ä–æ–±–æ—Ç–∞: ")
  
  currentX = 0
  currentY = 0

  historyOfPath = dict()
  historyOfPath[(0,0)] = 0 # dict –ø—É—Ç–µ–π, –≥–¥–µ = 0 - –¥–ª–∏–Ω–Ω–∞ –ø—Ä–æ–π–¥–µ–Ω–Ω–æ–≥–æ –ø—É—Ç–∏ –ø—Ä–∏ –¥–∞–Ω–Ω–æ–º "–¥–µ–π—Å—Ç–≤–∏–∏"

  bestStartInd = 0
  bestEndInd = n 

  wasFounded = False

  for i in range(n):
    #–≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç "–¥–µ–π—Å—Ç–≤–∏—è" —Ä–æ–±–æ—Ç–∞ –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –µ–≥–æ –Ω—ã–Ω–µ—à–Ω–µ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ
    if(s[i] == 'L'):
      currentX += -1
    elif(s[i] == 'R'):
      currentX += 1
    elif(s[i] == 'U'):
      currentY += 1
    elif(s[i] == 'D'):
      currentY += -1
     
    if ((currentX, currentY) in historyOfPath): #–µ—Å–ª–∏ —Ç–∞–∫–æ–µ –¥–µ–π—Å—Ç–≤–∏–µ —É–∂–µ –¥–æ —ç—Ç–æ–≥–æ –≤—Å—Ç—Ä–µ—á–∞–ª–æ—Å—å, —Ç–æ —É–∂–µ –Ω–µ -1 (—Ç–æ –µ—Å—Ç—å, –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø–æ–¥—Å—Ç—Ä–æ–∫–æ–π, —á—Ç–æ –º–æ–∂–Ω–æ —É–±—Ä–∞—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –≤—Å—è –ø–æ–¥—Å—Ç—Ä–æ–∫–∞)
      wasFounded = True
      if ((i+1) - historyOfPath[(currentX, currentY)] < bestEndInd - bestStartInd): #–µ—Å–ª–∏, —É–±—Ä–∞–≤ –∏–∑ –ø—É—Ç–∏ –¥–∞–Ω–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏–µ, –ø—É—Ç—å –º–æ–∂–µ—Ç —Å—Ç–∞—Ç—å –∫–æ—Ä–æ—á–µ
        #–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã —Ç–æ–≥–æ, —á—Ç–æ –º–æ–∂–Ω–æ —É–¥–∞–ª–∏—Ç—å
        bestEndInd = i + 1
        bestStartInd = historyOfPath[(currentX, currentY)] #–æ—Ç —Ç–æ—á–∫–∏, –±–µ–∑ –∫–æ—Ç–æ—Ä–æ–π –ø—É—Ç—å –º–æ–∂–µ—Ç –±—ã—Ç—å –∫–æ—Ä–æ—á–µ
    historyOfPath[(currentX, currentY)] = i + 1 #–¥–æ–±–∞–≤–ª—è–µ–º –¥–µ–π—Å—Ç–≤–∏–µ –∏ –Ω—ã–Ω–µ—à–Ω—é—é –¥–ª–∏–Ω—É –ø—É—Ç–∏ (–ø–æ—Å–ª–µ –¥–∞–Ω–Ω–æ–≥–æ –¥–µ–π—Å—Ç–≤–∏—è) –≤ –∏—Å—Ç–æ—Ä–∏—é –ø—É—Ç–∏

  if (wasFounded):
    print("–†–µ–∑—É–ª—å—Ç–∞—Ç: " + str(bestStartInd + 1) + " " + str(bestEndInd))
  else:
    print("–†–µ–∑—É–ª—å—Ç–∞—Ç: " + str(-1))  
```

–í–∞–º –∑–∞–¥–∞–Ω—ã —Ç—Ä–∏ —á–∏—Å–ª–∞:

```python
t = input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π: ")
t = int(t)

for i in range(t):

  lens = input("\n–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω—ã —Å—Ç—Ä–æ–∫: ")
  m, n = lens.split(" ")
  m, n = int(m), int(n)

  s = input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É: ")
  t = input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É: ")

  left_mas = [0 for i in range(n)]
  right_mas = [0 for i in range(n)]

  i = 0
  for j in range(len(t)): #–∏–¥—ë–º —Å–ª–µ–≤–∞ –¥–æ –∫–æ–Ω—Ü–∞ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ t
    while (s[i] != t[j]): #—É–≤–µ–ª–∏–≤–∞–µ–º –∏–Ω–¥–µ–∫—Å i –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –Ω–µ –ø—Ä–æ–∏–∑–æ–π–¥—ë—Ç —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ —Å —Å–∏–º–≤–æ–ª–æ–º —Å—Ç—Ä–æ–∫–∏ s
      i += 1
    left_mas[j] = i #—Å–æ—Ö—Ä–∞–Ω—è–µ–º, –Ω–∞ –∫–∞–∫–æ–º –∏–Ω–¥–µ–∫—Å–µ —Å–∏–º–≤–æ–ª —Å—Ç—Ä–æ–∫–∏ s —Å–æ–≤–ø–∞–ª —Å —Å–∏–º–≤–æ–ª–æ–º –∏–Ω–¥–µ–∫—Å–∞ —Å—Ç—Ä–æ–∫–∏ t
    i += 1

  i = len(s) - 1 #–¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å–∏–º–≤–æ–ª–æ–≤ —Å—Ç—Ä–æ–∫–∏ s –∏ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ t —Å –∫–æ–Ω—Ü–∞ 
  for j in range(len(t)-1, -1, -1): #–∏–¥—ë–º —Å –∫–æ–Ω—Ü–∞ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ t –¥–æ –Ω–∞—á–∞–ª–∞
    while (s[i] != t[j]):
      i -= 1
    right_mas[j] = i
    i -= 1

  final = 0
  for i in range(n - 1): #–∏–¥—ë–º –ø–æ –¥–ª–∏–Ω–Ω–µ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ –∏ –Ω–∞—Ö–æ–¥–∏–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é —à–∏—Ä–∏–Ω—É 
    if (final < right_mas[i + 1] - left_mas[i]):
      final = right_mas[i + 1] - left_mas[i]

  print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —à–∏—Ä–∏–Ω–∞ —Ö–æ—Ä–æ—à–µ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏: {final}")
```
