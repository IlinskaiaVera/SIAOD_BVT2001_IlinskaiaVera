# Задание 1

Даны две строки: `s1` и `s2` с одинаковым размером.

Проверьте, может ли некоторая перестановка строки `s1` “победить” некоторую перестановку строки `s2` или наоборот.

Строка x может “победить” строку y (обе имеют размер n), если `x[i] >= y[i]` (в алфавитном порядке) для всех i от 0 до n-1.


## Ввод

`abc`
`xya`

## Вывод

`True`

### Исполняемый код записать в ячейку ниже:

```python
s1 = input("Введите s1: ")
s2 = input("Введите s2: ")

#сортировка так, чтобы шли по буквенному алфавиту
s1_sorted = ''.join(sorted(s1))
s2_sorted = ''.join(sorted(s2))

win_s1 = True
win_s2 = True

#выигрывает s1 или s2, если все их буквы
#больше букв другой строки
for i in range(len(s1)):
    if (ord(s1_sorted[i]) > ord(s2_sorted[i])):
        win_s2 = False #s2 уже не может выиграть
    if (ord(s2_sorted[i]) > ord(s1_sorted[i])):
        win_s1 = False #s1 уже не может выиграть
        
result = bool(win_s1 + win_s2) #если хоть один True, значит может победить либо s1, либо s2
print(result)
```

# Задание 2

Дана строка `s`, вернуть самую длинную полиндромную подстроку в `s`.

## Ввод

`babad`

## Вывод

`aba` или `bab`

## Исполняемый код записывать в ячейку ниже

```python
s = input("Введите полиндромную строку: ")
max_poli = ""

for i in range (0, len(s)-1): #первая позиция подстроки
    for j in range (i+1, len(s)): #последняя позиция подстроки
        pod_str = s[i:j] #берём подстроку
        #сравниваем подстроку с ревёрсной подстрокой
        if (pod_str == pod_str[::-1] and len(max_poli) < len(pod_str)):
            max_poli = pod_str
        #если сама строка, а не подстрока в ней 
        #является полиндромом и больше найденного max_poli
        if (s == s[::-1] and len(max_poli)<len(s)):
            max_poli = s
print ("Самый длинный полиндром: " + max_poli)
```

# Задание 3

Вернуть количество отдельных непустых подстрок текста, которые могут быть записаны как конкатенация некоторой строки с самой собой (т.е. она может быть записана, как `a + a`, где `a` - некоторая строка).

## Ввод

`aabb`

## Вывод

2

## Исполняемый код записывать в ячейку ниже

```python
s = input("Введите строку: ")

result = 0

for i in range (0, len(s)-1): 
    for j in range (i+1, len(s)):
        if(s[i:j] == s[(i+1):(j+1)]):
           result += 1
print(result)
```
