####Кривая Коха

```python
iterations = input("Введите глубину фрактала: ")
iterations_n = int(iterations)
size = 300 #длина линии

initializeTurtle()
turtle.right(90) #повернули черепашку, так как изначально она почему-то идёт вверх
#ниже переместим черепашку немного налево, чтобы ближе к серединке рисовалась кривая
turtle.penup()
turtle.backward(180)
turtle.pendown()

turtle.speed(10) #изменим скорость рисования черепашки

def kohCurve(iterations_n, size):
  if (iterations_n == 0):
    turtle.forward(size)
  else:
    kohCurve(iterations_n-1, size/3)
    turtle.left(60) #идём наверх
    kohCurve(iterations_n-1, size/3)
    turtle.right(120) #идём вниз (60+60)
    kohCurve(iterations_n-1, size/3)
    turtle.left(60) #идём по прямой
    kohCurve(iterations_n-1, size/3)

start_time = time.perf_counter() 

kohCurve(iterations_n, size)

final_time = time.perf_counter()
kohCurveTime = (final_time-start_time)/0.10
```

####Кривая Гильберта

```python
iterations = input("Введите глубину фрактала: ")
iterations_n = int(iterations)
size = 20 #длина линии
ang = 90 #в данном случае потребуется изменять угол с 90 на -90

initializeTurtle()
turtle.right(90) #повернули черепашку, так как изначально она почему-то идёт вверх
#ниже переместим черепашку немного налево и наверх
turtle.penup()
turtle.backward(180)
turtle.left(90)
turtle.forward(90)
turtle.right(90)
turtle.pendown()

turtle.speed(10) #изменим скорость рисования черепашки

def hilbertCurve(iterations_n, size, ang):
  if (iterations_n == 0):
    return
  else:
    turtle.right(ang) #при первой итерации повернули черепашку вправо, дабы она пошла вниз
    hilbertCurve(iterations_n-1, size, -ang)
    turtle.forward(size)
    turtle.left(ang) #теперь влево, дабы она сделала линию вправо
    hilbertCurve(iterations_n-1, size, ang)
    turtle.forward(size)
    hilbertCurve(iterations_n-1, size, ang)
    turtle.left(ang) #снова влево, дабы черепашка пошла наверх
    turtle.forward(size)
    hilbertCurve(iterations_n-1, size, -ang)
    turtle.right(ang)

start_time = time.perf_counter() 

hilbertCurve(iterations_n, size, ang)

final_time = time.perf_counter()
hilbertCurveTime = (final_time-start_time)/0.10
```

####Кривая Серпинского

```python
iterations = input("Введите глубину фрактала: ")
iterations_n = int(iterations)
size = 20 #длина линии

initializeTurtle()
turtle.right(90) #повернули черепашку, так как изначально она почему-то идёт вверх
#ниже переместим черепашку немного налево, чтобы ближе к серединке рисовалась кривая
turtle.penup()
turtle.backward(180)
turtle.pendown()

turtle.speed(10) #изменим скорость рисования черепашки

#рисуем "клевер" (или диагональ клевера, если будет 1 итерация)
def polu_serp_curve(iterations_n, size):
  if (iterations_n == 0):
    return
  polu_serp_curve(iterations_n - 1, size)
  turtle.left(45)
  turtle.forward(size*(math.sqrt(2)))
  turtle.left(45)
  polu_serp_curve(iterations_n - 1, size)
  turtle.right(90)
  turtle.forward(size)
  turtle.right(90)
  polu_serp_curve(iterations_n - 1, size)
  turtle.left(45)
  turtle.forward(size*(math.sqrt(2)))
  turtle.left(45)
  polu_serp_curve(iterations_n - 1, size)

#рисование всей кривой
def all_serp_curve(iterations_n, size):
  polu_serp_curve(iterations_n, size)
  turtle.right(90)
  turtle.forward(size)
  turtle.right(90)
  polu_serp_curve(iterations_n, size)
  turtle.right(90)
  turtle.forward(size)
  turtle.right(90)

start_time = time.perf_counter() 

all_serp_curve(iterations_n, size)

final_time = time.perf_counter()
serpCurveTime = (final_time-start_time)/0.10
```

####Ковёр Серпинского

```python
iterations = input("Введите глубину фрактала: ")
iterations_n = int(iterations)
size = 150 #длина линии

initializeTurtle()
turtle.right(90) #повернули черепашку, так как изначально она почему-то идёт вверх
#ниже переместим черепашку немного налево, чтобы ближе к серединке рисовалась кривая
turtle.penup()
turtle.backward(180)
turtle.pendown()

turtle.speed(10) #изменим скорость рисования черепашки
    
def сarpetSer(iterations_n, size):
  if (iterations_n == 0):
    #рисуем последний квадратик
    for _ in range(4): 
        turtle.forward(size)
        turtle.left(90)
  else:
    for _ in range(4):
        сarpetSer(iterations_n - 1, size/3)
        turtle.forward(size/3)
        сarpetSer(iterations_n - 1, size/3)
        turtle.forward(size/3)
        turtle.forward(size/3)
        turtle.left(90)
 
start_time = time.perf_counter() 

сarpetSer(iterations_n, size)

final_time = time.perf_counter()
carpetSerTime = (final_time-start_time)/0.10
```

####Дерево Пифагора

```python
import math

iterations = input("Введите глубину фрактала: ")
iterations_n = int(iterations)
size = 40 #длина линии

initializeTurtle()
turtle.right(90) #повернули черепашку, так как изначально она почему-то идёт вверх

turtle.speed(7) #изменим скорость рисования черепашки

#функция для рисования Пифагоровых штанов, из которых и образуется дерево Пифагора
def pyth_pants(iterations_n, size):
  if (iterations_n <= 0):
    turtle.forward(size) #дорисовываем закрывающую линию
  else:
    turtle.right(135)
    turtle.forward(((size*size)/2)**(0.5)) #от sqrt всё рисовалось не так, степень пришлось писать иначе
    turtle.left(90)
    pyth_pants(iterations_n-1, ((size*size)/2)**(0.5))
    turtle.left(90)
    turtle.forward(((size*size)/2)**(0.5))
    turtle.left(90)
    turtle.forward(((size*size)/2)**(0.5))
    turtle.right(135)
    turtle.forward(size)
    turtle.right(135)
    turtle.forward((size*size/2)**(0.5))
    turtle.left(90)
    turtle.forward((size*size/2)**(0.5))
    turtle.left(90)
    pyth_pants(iterations_n-1, ((size*size)/2)**(0.5))
    turtle.left(90)
    turtle.forward((size*size/2)**(0.5))
    turtle.right(135)

def pyth_tree(iterations_n, size):
  turtle.forward(size)
  turtle.left(90)
  turtle.forward(size)
  turtle.left(90)
  pyth_pants(iterations_n, size)
  turtle.left(90)
  turtle.forward(size)

start_time = time.perf_counter() 

pyth_tree(iterations_n, size)

final_time = time.perf_counter()
pythTreeTime = (final_time-start_time)/0.10
```
