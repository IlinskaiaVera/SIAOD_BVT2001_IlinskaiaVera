# Задание №3
Методы сортировки строк случайно матрицы

#### Функция сортировки выбором

```python
#сравнение и обмен фиксированного a элемента с элементами от a+1 до длинны строки
def Selection_Sort (arr1):
    for i in range (0, len(arr1)): # len(arr) == m
         for a in range (0, len(arr1[i])-1): # len(arr[i]) == n
            for b in range(a+1, len(arr1[i])): 
                if (arr1[i][a] > arr1[i][b]):
                    arr1[i][a], arr1[i][b] = arr1[i][b], arr1[i][a]
    return arr1

arr_selection = Selection_Sort(arr)
for s in arr_selection:
    print(s)
```

#### Функция сортировки вставкой

```python
#сравнение и обмен фиксированного а элемента с элементами a-1 (справа налево)
def Insertion_Sort (arr2):
    for i in range(0, len(arr2)):
        for a in range(0, len(arr2[i])):
            fics = arr2[i][a]
            pos = a
            while (pos>0 and arr2[i][pos-1] > fics):
                arr2[i][pos] = arr2[i][pos-1]
                pos-=1
                arr2[i][pos] = fics
    return arr2

arr_insertion = Insertion_Sort(arr)
for s in arr_insertion:
    print(s)
```

#### Функция сортировки обменом (пузырьком)

```python
#сравнение пар соседних элементов и обмен
def Exchange_Sort (arr3):
    for i in range(0, len(arr3)):
        for run in range(0, len(arr3[i])-1): #для многоразовой сортировки одной строки по парам
            for a in range (len(arr3[i])-1): #len-1, так как у последнего элемента нету пары a+1
                if (arr3[i][a] > arr3[i][a+1]): #пузырьком является последнее число обмена
                    arr3[i][a], arr3[i][a+1] = arr3[i][a+1], arr3[i][a]
    return arr3

arr_exchange = Exchange_Sort(arr)
for s in arr_exchange:
    print(s)
```

#### Функция сортировки Шелла (улучшенная сортировка вставкой)

```python
#сортировка подмассивов и сравнение элементов через определённый диапазон
def Shell_Sort (arr4):
    for i in range (0, len(arr4)):
        m = len(arr4[i])
        interval = m//2 #с помощью интвервала разделяем массив на подмассивы
                        #их элементы мы будем между собой сравнивать
        while (interval > 0):
            for a in range(interval, m):
                fics = arr4[i][a] #берём элемент второго подмассива
                pos = a #сохраняем позицию элемента второго подмассива
                while (pos >= interval and arr4[i][pos-interval] > fics):
                    #arr[i][a-interval] > fics - сравнение элемента первого массива со вторым 
                    arr4[i][pos] = arr4[i][pos-interval]
                    pos -= interval
                    arr4[i][pos] = fics
            interval //= 2 #например, если 8, то разбиваем на 4, затем на 2 и 1, и сравниваем элементы
    return arr4

arr_shell = Shell_Sort(arr)
for s in arr_shell:
    print(s)
```

#### Функция быстрой сортировки

```python
#сортировка, основанная на рекурсии и выборкой опорного
#элемента на каждом рекурсивном этапе


def Fast_Sort (str5):
    less = []
    equal = [] 
    more = []
    if (len(str5) > 1):
        opora = str5[0]
        for x in str5:
            if (x < opora):
                less.append(x)
            elif (x == opora):
                equal.append(x)
            elif (x > opora):
                more.append(x)
        return Fast_Sort(less) + equal + Fast_Sort(more)
    else:
        return str5

final_fastsort_list5 = [];

for i in range(len(arr)):
    final_fastsort_list5.append(Fast_Sort(arr[i]))
for j in range (len(final_fastsort_list5)):
    print(final_fastsort_list5[j])
```

#### Функция пирамидальной сортировки

```python
#сортировка за счёт создания "родственных связей" и сравнения элементов
#в первый раз от потомка до предка, в последующие - от предка к потомку

#главная функция
def Pyramid_Sort(str6, a, b):
    #позиция самого большого элемента, с которого 
    #по "родственным связям" начинается сравнение - супер предка
    pos_large = b
        
    #позиция крайнего левого предка
    pos_left_desce = 2*b + 1
    #позиция крайнего правого предка
    pos_right_desce = 2*b + 2
        
    #проверка для крайнего левого потомка
    #проверка на допустимость индекса позиции и на то, является ли элемент больше супер предка
    if (pos_left_desce < a and str6[pos_left_desce] > str6[pos_large]):
        pos_large = pos_left_desce

    #проверка для крайнего правого потомка 
    if (pos_right_desce < a and str6[pos_right_desce] > str6[pos_large]):
        pos_large = pos_right_desce

    #если наибольший элемент уже не корневой, то меняются местами
    if pos_large != b:
        str6[b], str6[pos_large] = str6[pos_large], str6[b]
        Pyramid_Sort(str6, a, pos_large)
        
#вспомогаетельная функция для функции пирамидальной сортировки
#чтобы использовать на матрице и сделать сортировку
def Auxiliary_Sort(arr):  
    a = len(arr)
    
    #остановка алгоритма перед элементом -1 -
    #перед первым элементом списка
    #ещё один -1 означает повторный проход по списку 
    #в обратном направлении, уменьшая счётчик a на 1 
    for b in range(a, -1, -1):
        Pyramid_Sort(arr, a, b)

    #перемещаем супер предка в конец списка
    for b in range(a-1, 0, -1):
        arr[b], arr[0] = arr[0], arr[b]
        Pyramid_Sort(arr, b, 0)
    return arr

final_pyramidsort_list6 = [];

for i in range(len(arr)):
    final_pyramidsort_list6.append(Auxiliary_Sort(arr[i]))
for j in range (len(final_pyramidsort_list6)):
    print(final_pyramidsort_list6[j])
```

#### Функция турнирной сортировки

```python
#сортировка, в основе которой лежит создание двух ветвей и их сравнение
# с элементами массива для определения "победителей" и "проигравших"

def Tournament_Sort(str7):
    temp_mas = [None] * 2 * (len(str7) + len(str7)%2)
    j = len(temp_mas) - len(str7) - len(str7)%2
    
    for index, element in enumerate(str7):
        temp_mas[j + index] = (index, element)
    
    for i in range(len(str7)):
        n = len(str7)
        j = len(temp_mas) - len(str7) - len(str7)%2
        while j > -1:
            n = (n + 1)//2
            for index in range(n):
                index = max(j + index*2, 1)
                if temp_mas[index] is not None and temp_mas[index + 1] is not None:
                    temp_mas[index//2] = temp_mas[index + (temp_mas[index][1]
                                                         >= temp_mas[index + 1][1])]
                else:
                    temp_mas[index//2] = temp_mas[index] if temp_mas[index] is not None else temp_mas[index + 1]
            j -= n
            
        j, x = temp_mas[0] #значения из двух веток кладём в переменные
        str7[i] = x
        temp_mas[len(temp_mas) - len(str7) - len(str7)%2 + j] = None
    return str7

final_tournamentsort_list7 = []

for i in range(len(arr)):
    final_tournamentsort_list7.append(Tournament_Sort(arr[i]))
for j in range (len(final_tournamentsort_list7)):
    print(final_tournamentsort_list7[j])
```
